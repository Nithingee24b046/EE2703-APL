```python
import numpy as np
x,y=np.loadtxt('dataset1.txt', unpack=True)
```

```python
%matplotlib ipympl
import matplotlib.pyplot as plt
```

```python
plt.plot(x, y)
```

```python
len(x)
```

```python
A = np.vstack([x, np.ones(len(x))]).T

# Use lstsq to solve for the coefficients [m, b]
coefficients, residuals, rank, s = np.linalg.lstsq(A, y, rcond=None)

# Extract slope and intercept
slope, intercept = coefficients

print(f"Linear fit: y = {slope:.4f}x + {intercept:.4f}")
print(f"Slope: {slope:.4f}")
print(f"Intercept: {intercept:.4f}")
```

```python
# Calculate R-squared for goodness of fit
y_pred = slope * x + intercept
ss_res = np.sum((y - y_pred) ** 2)
ss_tot = np.sum((y - np.mean(y)) ** 2)
r_squared = 1 - (ss_res / ss_tot)
print(f"R-squared: {r_squared:.4f}")

# Plot the data and fitted line
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.7, label='Data points')
plt.plot(x, y_pred, color='red', linewidth=2, label=f'Fitted line: y = {slope:.2f}x + {intercept:.2f}')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Linear Fit using NumPy lstsq')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Optional: Print some statistics about the fit
print(f"\nFit statistics:")
print(f"Mean absolute error: {np.mean(np.abs(y - y_pred)):.4f}")
print(f"Root mean square error: {np.sqrt(np.mean((y - y_pred)**2)):.4f}")
```
