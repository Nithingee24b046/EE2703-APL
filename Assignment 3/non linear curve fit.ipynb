```python
# Imports and settings
import numpy as np
import matplotlib.pyplot as plt
%matplotlib ipympl
```

```python
# Lambda Expressions in Python: Examples for Curve Fitting

def sq(x):
    return x**2

# Basic lambda expression
square = lambda x: x**2
print(square(4))  # Output: 16
```

```python
# Using lambda in curve fitting scenarios

import numpy as np
from scipy.optimize import curve_fit
```

```python
# Example 1: Simple linear function
x_data = np.array([0, 1, 2, 3, 4, 5])
y_data = np.array([1, 3, 5, 7, 9, 11])

# Traditional function definition
def linear_func(x, m, b):
    return m*x + b

# Equivalent lambda expression
linear_lambda = lambda x, m, b: m*x + b

# Using lambda in curve_fit
popt, _ = curve_fit(linear_lambda, x_data, y_data)
print("Fitted parameters:", popt)
plt.figure(); plt.plot(x_data, y_data, x_data, linear_lambda(x_data, *popt) )
```

```python
# Example 2: Exponential decay
x_data = np.linspace(0, 5, 50)
y_data = 3 * np.exp(-1.2 * x_data) + np.random.normal(0, 0.1, 50)

exp_decay = lambda x, a, b: a * np.exp(-b * x)
popt, _ = curve_fit(exp_decay, x_data, y_data)
print("Fitted parameters for exp decay:", popt)
plt.figure(); plt.plot(x_data, y_data, x_data, exp_decay(x_data, *popt))
```

```python
# Example 3: Using lambda for on-the-fly function creation
def create_poly_func(degree):
    return lambda x, *params: sum(params[i] * x**i for i in range(degree+1))

quadratic = create_poly_func(2)
x_data = np.linspace(0, 5, 100)
y_data = quadratic(x_data, 1, 2, 1) + np.random.randn(len(x_data))

# plt.figure(); plt.plot(x_data, y_data)
popt, _ = curve_fit(quadratic, x_data, y_data, [0, 0, 0])
print("Fitted parameters for quadratic:", popt)
plt.figure(); plt.plot(x_data, y_data, x_data, quadratic(x_data, *popt))
```

```python
# Example 4: sine wave sum
def sine_sum(x, A, f):
    return A * np.sin(2*np.pi*f*x)

x_data = np.linspace(0, 5, 100)
y_data = sine_sum(x_data, 1, 0.2) + 0.01*np.random.randn(len(x_data))

# plt.figure(); plt.plot(x_data, y_data)
popt, _ = curve_fit(sine_sum, x_data, y_data, [0.1, 0])
print("Fitted parameters for sine_sum:", popt)
plt.figure(); plt.plot(x_data, y_data, x_data, sine_sum(x_data, *popt))
# plt.figure(); plt.plot(x_data, y_data)
```
